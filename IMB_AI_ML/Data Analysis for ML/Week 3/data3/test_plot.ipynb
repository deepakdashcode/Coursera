```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('own_data.csv')
```

```python
df
```

```python
df['Price'] = df['Price'].apply(lambda x: round(x))
```

```python
df
```

```python
plt.plot(df.Price, df.Rating, ls='', marker='o', label='Rating')
plt.plot(df.Price, df.Name, ls='', marker='o', label='Name')
```

```python
plt.hist(df.Price, bins=10)
```
